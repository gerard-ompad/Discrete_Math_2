### Task 1:
Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

**Question 1:** Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

**Solution:** <br>

$E[y] = \sum_{i=1}^{k} y_i \cdot P(y_i) = \sum_{i=1}^{k} y_i \cdot \left(\frac{n!}{y_1!y_2! \ldots y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot \ldots \cdot p_k^{y_k}\right)$ <br><br>

Since $(Y_1 + Y_2 + \ldots + Y_k = n)$, we can express  $(y_i)$ as $(n - (Y_1 + Y_2 + \ldots + Y_{i-1} + Y_{i+1} + \ldots + Y_k))$:

$E[y] = \sum_{i=1}^{k} \left(n - \sum_{j \neq i} y_j\right) \cdot \left(\frac{n!}{y_1!y_2! \ldots y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot \ldots \cdot p_k^{y_k}\right)$

Distribute the summation and simplify:

$E[y] = n \cdot \sum_{i=1}^{k} \left(\frac{(n-1)!}{y_1!y_2! \ldots y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot \ldots \cdot p_k^{y_k}\right)$

The term inside the summation is the joint probability distribution for (n-1) trials with probabilities $(p_1, p_2, \ldots, p_k)$.
Therefore, the summation evaluates to 1:

$E[y] = n \cdot 1 = n$

We’ve shown that $E[y] = n \cdot p_i$, where $(p_i)$ is the probability of an outcome falling into cell $(i)$.
Thus, the expected value of the random variable $(y)$ is indeed $(n \cdot p_i)$. <br><br>

**Solution:**

The expected value $E[X]$  for selecting a person aged 65 and over from the sample is determined by the formula $E[X] = n \cdot p$, where $ n$ represents the sample size and $p$ is the probability of choosing someone from the 65 and above age group.
<br><br>
For age 18-24: $$P(X=100) = \binom{500}{100} \cdot (0.18)^{100} \cdot (0.82)^{400}$$ <br><br>

Therefore, the expected value for to obtain a person in the 65 and above is 80 <br><br>

### Task 2

Consider the following experiment:

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

**Question 1:** From the following steps shown above, derive the probability distribution of the experiment.

**Solution:**

The probability distribution of a binomial experiment is given by the binomial probability formula: <br>

$$P(Y = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k}$$ <br>
where, $k$ is the number of successess(the value that $Y$ takes). $\binom{n}{k}$ represents the number of ways to choose $k$ successes out of $n$ trials <br><br>

**Question 2:** Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.

**Solution:**

Using the probability distribution:

$$P(Y = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k}$$ <br><br>

The expectation $E[Y]$ is calculated as:

$$E[Y] = \sum_{k=0}^{n} y \cdot P(Y = y)$$ <br><br>

Substituting the PMF of the binomial distribution:

$$E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k}$$ <br><br>

We can use the identity $k \cdot \binom{n}{k} = n \cdot \binom{n-1}{k-1}$ to simplify this expression.This identity states that choosing $k$
items out of $n$ and then picking one out of these $k$ as distinct can be equated to picking one out of $n$ as distinct and then choosing $k-1$
from the remaining $n-1$.

So, the expression becomes:

$$E[Y] = \sum_{k=0}^{n} n \cdot \binom{n-1}{k-1} \cdot p^{k} \cdot (1-p)^{n-k}$$

Now, notice that this expression resembles the binomial expansion of $(p + (1-p))^{n}$, which is $1^{n} = 1$.
Therefore, the sum simplifies to $n \cdot 1 = n$.

Thus, $$E[Y] = n \cdot p$$<br><br>

**Question 3:** Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

**Solution:**

Using the probability distribution:

$$P(Y = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k}$$ <br><br>

where:
$n$ is the number of trials (10 in this case)
$k$ is the number of successes (in this case, either 9 or 10)
$p$ is the probability of success (0.30 in this case)
$(1-p)$ is the probability of failure

Using this formula:

For $k = 9$:<br><br>
$$P(Y = 9) = \binom{10}{9} \cdot (0.30)^{9} \cdot (1 - 0.30)^{10-9}$$ <br><br>

For $k = 10$:<br><br>
$$P(Y = 10) = \binom{10}{10} \cdot (0.30)^{10} \cdot (1 - 0.30)^{10-10}$$ <br><br>

Then, the probability of at least nine people recovering is the sum of these probabilities:

$$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$$

Now, let's calculate each part:

For $k = 9$: <br><br>
$$P(Y = 9) = \binom{10}{9} \cdot (0.30)^{9} \cdot (1 - 0.30)^{10-9}$$ <br><br>
$$P(Y = 9) = \binom{10}{9} \cdot (0.30)^{9} \cdot (0.70)^{1}$$ <br><br>
$$P(Y = 9) = 10 \cdot (0.30)^{9} \cdot 0.70$$ <br><br>
$$P(Y = 9) \approx 0.006006$$ <br><br>

For $k = 10$: <br><br>
$$P(Y = 10) = \binom{10}{10} \cdot (0.30)^{10} \cdot (1 - 0.30)^{10-10}$$ <br><br>
$$P(Y = 10) = \binom{10}{10} \cdot (0.30)^{10} \cdot (0.70)^{0}$$ <br><br>
$$P(Y = 10) = 1 \cdot (0.30)^{10} \cdot 1$$ <br><br>
$$P(Y = 10) \approx 0.000005904$$ <br><br>

Now, sum these probabilities:

$$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$$ <br><br>
$$P(Y \geq 9) \approx 0.006006 + 0.000005904$$ <br><br>
$$P(Y \geq 9) \approx 0.006011904$$ <br>

So, the probability that at least nine out of ten people receiving the medication will recover, given that the medication is worthless, is approximately $0.006011904$.<br><br>


**Question 4:** Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

**Solution:**

We'll use the binomial distribution formula to calculate the probability of observing exactly $k$ defectives in a sample of size $n$ from a population with proportion $p$ of defectives:

$$P(X = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k}$$ <br><br>

where:

$n = 5$ (sample size) <br><br>
$p = 0.05$ (proportion of defectives in the lot)<br><br>
$k = 0$(no defective fuses observed)<br><br>

Calculate for no defective fuses in a sample of 5:

$P(X = 0) = \binom{5}{0} \cdot 0.05^{0} \cdot (1 - 0.05)^{5}$ <br><br>
$ = \binom{5}{0} \cdot 1 \cdot 0.95^{5}$ <br><br>
$ = 1 \cdot 0.95^{5}$ <br><br>
$ \approx 0.77378$ <br><br>

Now, the probability of observing at least one defective fuse:

$P$ (at least one defective) $ = 1 - P$ (no defective)<br><br>
$ = 1 - 0.77378$ <br><br>
$ \approx 0.22622$ <br><br>

Therefore, the probability of observing at least one defective fuse in a sample of 5 fuses is approximately 0.22622 or 22.622%.<br><br>

## Task 3:

Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

<b>Question 1:</b> Find a general formula for the expected value of this distribution.

<b>Solution:</b>

$$E[y] = \sum_{y =0}^{\infty} y \cdot p(y)$$ <br><br>
$$ = \sum_{y =0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$ <br><br>
$$ = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!}$$ <br><br>
$$ = e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!}$$ <br><br>
$$ = e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}$$ <br><br>

Now, notice that the sum is just the Taylor series expansion of $e^\lambda$ which is $\sum_{n=0}^{\infty} \frac{\lambda^{n}}{n!}$.
Using this, we can rewrite the expression for $E[y]$ as:

$$E[y] = e^{-\lambda} \cdot (\lambda \cdot e^{\lambda})$$ <br><br>
$$E[y] = \lambda$$<br><br>

So, the general formula for the expected value of the given distribution is $E[y] = \lambda$.<br><br>

<b>Question 2:</b> A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

<b>Solution:</b>

To find the probability that none of the ten 1-square-yard sampling regions will contain seedlings, we can use the Poisson distribution, assuming that the distribution of seedlings follows a Poisson distribution with a mean density of five per square yard.

The Poisson probability mass function is given by:

$$P(X = k) = \frac{e^{-\lambda}\lambda^{k}}{k!}$$

where:

$P(X = k)$ is the probability of observing $k$  seedlings in a given region. <br><br>
$\lambda$ is the mean number of seedlings per region. <br><br>
$e$ is Euler's number, approximately equal to 2.71828. <br><br>
$k!$ is the factorial of $k$.<br><br>

Given that the mean density of seedlings is five per square yard, the mean number of seedlings per 1-square-yard region $(\lambda)$ is also five.

We want to find the probability that none of the regions will contain seedlings, which means $k = 0$.

So, plugging the values into the Poisson probability mass function:

$$P(X = 0) = \frac{e^{-5}\,\cdot\,5^{0}}{0!}$$ <br><br>
$$P(X = 0) = e^{-5}$$ <br><br>

Now, we can calculate this probability:

$$P(X = 0) \approx e^{-5} \approx 0.0067$$

So, the probability that none of the ten 1-square-yard sampling regions will contain seedlings is approximately <b>0.0067</b>, or about <b>0.67%.</b>
